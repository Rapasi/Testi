```python
vastaukset=religion.groupby('How comfortable would you be seeing someone who practices a different religion from you: Tell someone \"I\'ll pray for you\"').size()
vastaukset[0] = np.roll(vastaukset[0], shift=5)
```

```python
a='Extremely comfortable'
b='Not at all comfortable'
c='Not so comfortable    '  
d='Somewhat comfortably '
e='Very comfortable'
bins=[a,b,c,e,d]
plt.figure(figsize=(24,15))
bars=plt.bar(bins,vastaukset)
plt.title('How comfortable would you be seeing someone who practices a different religion from you: Tell someone \"I\'ll pray for you?\"',size=30)
koko = 20
plt.rc('xtick', labelsize=koko)   
plt.rc('ytick', labelsize=koko)
pattern=['/','*','O','|','+']
for bar in bars:
    bar.set_hatch(pattern.pop(0))
```
